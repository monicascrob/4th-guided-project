# Students' job outcomes 

In this project we are exploring the job outcomes for students who graduated from college between 2010 and 2012 in US. The dataset we are using is a cleanded dataset released by FiveThirtyEight from a saataset realeased by American Community Survey, which conducts surveys and aggregates data.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries and more.

These are some of the columns in our dataset:
 - Rank - by median earnings
 - Major_code
 - Major
 - Major_category
 - Total
 - Sample_size
 - Men
 - Women
 - ShareWomen
 - Employed
 - Median
 - Low_wage_jobs
 - Full_time
 - Part_time
 
Using visualization methods (scatter plots, histograms, bar plots) we will try to answer the following questions:


 - Do students in more popular majors make more money?
 - How many majors are predominantly male? Predominantly female?
 - Which category of majors have the most students?
